```python
import cv2
import urllib.request
from matplotlib import pyplot as plt
import numpy as np
```

```python
urllib.request.urlretrieve(url, 'toto.jpg')
```

```python
toto = cv2.imread('toto.jpg')
toto_correct = cv2.cvtColor(toto, cv2.COLOR_BGR2GRAY)
```

```python

plt.hist(toto_correct.ravel(),256,[0,256])
plt.title('Histogram of Grayscale trumptrudeau.jpg')
plt.show()
```

```python

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([toto],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
